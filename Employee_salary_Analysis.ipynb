Which country has highest salary in Data science?

```python
details=df.sort_values(by="salary",ascending=False)[["company_location","salary"]].head(5)
```

```python
company_locations = details["company_location"]
salaries = details["salary"]

plt.figure(figsize=(10, 6))
plt.bar(company_locations, salaries, color='purple')
plt.xlabel('Company Location')
plt.ylabel('Salary')
plt.title('Top 5 Salaries for Data scientist by Company Location')
plt.xticks(rotation=45)
plt.show()
```

*Chile offers more  jobs compared to other countries*

Top 5 jobroles which has highest salary vs company_location

```python
job_roles=df.sort_values(by="salary",ascending=False)[["job_title","company_location","salary"]].head(5)
job_roles
```

```python


# Sample data from your DataFrame
job_titles = job_roles["job_title"]
company_locations = job_roles["company_location"]
salaries = job_roles["salary"]

# Map unique values to numerical indices for plotting
location_codes = {location: i for i, location in enumerate(company_locations.unique())}
job_codes = {job: i for i, job in enumerate(job_titles.unique())}

# Convert categorical data to numerical data for plotting
x = [location_codes[loc] for loc in company_locations]
y = [job_codes[job] for job in job_titles]
sizes = salaries / 10000  # Adjust size of bubbles for better visualization

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=sizes, alpha=0.6, color='black', edgecolor='w', linewidth=1)

# Customize axis labels
plt.xticks(ticks=range(len(location_codes)), labels=location_codes.keys(), rotation=45)
plt.yticks(ticks=range(len(job_codes)), labels=job_codes.keys())

plt.xlabel('Company Location')
plt.ylabel('Job Title')
plt.title('Salary Distribution by Job Title and Company Location')


plt.show()
```

*Data states that DataScientist in Chile have highest salary compare to others*

*Top 5 job that offers highest salary for Entry level employees*

```python
top_5=df.loc[df["experience_level"]=="EN"].sort_values(by="salary",ascending=False)[["job_title","salary"]].head(5)
top_5
```

```python
sorted_emp = df.groupby('employment_type')['salary_in_usd'].mean().sort_values(ascending=True).index
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
ax = sns.countplot(x='employment_type', data=df, order=sorted_emp)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'),
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center',
                xytext = (1, 7),
                textcoords = 'offset points',
               fontsize=16)

plt.xlabel('Employment Type')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=0, ha="right")
plt.title('Number of Jobs vs Employment Type')


plt.tight_layout()
plt.show()
```

*Number of jobs for full time employee is highest compare to another employment type*

```python
avg_salary_df = df.groupby('employment_type')['salary_in_usd'].mean()

# Plot the graph
avg_salary_df.plot(kind='bar', color='purple')
plt.title('Employment Type vs Average Salary')
plt.xlabel('Employment Type')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.show()
```

Contract employee has highest salary as compare to others

```python
data_scientist_df = df[df['job_title'] == 'Data Scientist']

# Group by year and calculate the average salary
avg_salary_by_year = data_scientist_df.groupby('work_year')['salary_in_usd'].mean()

# Plot the graph
avg_salary_by_year.plot(kind='line', marker='o', color='purple')
plt.title('Data Scientist Salary vs Work Year')
plt.xlabel('Work Year')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

*The trend for Data Scientist roles is expected to increase as both the demand for the position and the associated salaries continue to rise.*

```python
entry_level_df = df[df['experience_level'] == 'EN']

# Group by employment type and company size
grouped_df = entry_level_df.groupby(['employment_type', 'company_size']).size().unstack()

# Plot the graph
grouped_df.plot(kind='bar', stacked=False)
plt.title('Entry Level Employment Type vs Company Size')
plt.xlabel('Employment Type')
plt.ylabel('Number of Companies')
plt.show()
```

* Full-time job offers from larger companies tend to be higher in terms of salary and compensation. Additionally, larger companies often have more full-time positions available compared to smaller organizations*
